```python
#import libraries
from bs4 import BeautifulSoup
import requests
import os
```

```python
#identify url we will be scraping
url = "https://www.birthcontrol.com/options/iud/paragard/"
```

```python
#access data within url
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
html = soup.html
```

```python
print(soup.prettify())
```

```python
bc_type = soup.find('div', attrs = {'class': 'et_pb_text_inner'}).find('h1').text
bc_type
```

```python
list_symptoms = soup.find('div', attrs = {'class': 'et_pb_text_inner'}).find('ul')
list_symptoms
```
