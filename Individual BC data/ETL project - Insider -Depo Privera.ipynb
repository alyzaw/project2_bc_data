```python
#import libraries
from bs4 import BeautifulSoup
import requests
import os
```

```python
#identify url we will be scraping
url = "https://www.rxlist.com/depo-provera-drug.htm#description"
```

```python
#access data within url
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
html = soup.html
```

```python
print(soup.prettify())
```

```python
bc_type = soup.find('section').find('h1').text
bc_type
```

```python
list_symptoms = soup.find('body', class_= 'rxArt pgprof').find('main', class_= 'content').find('section', class_= 'story').find('article', attrs = {'class': 'pageContainer'}).find('section', class_='pageSection')[3]#.find('div', class_= 'apPage'). find('div', class_='wrapper').find('div', class_= 'pgContent').find('h3')
list_symptoms
```
