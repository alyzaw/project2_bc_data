```python
#import libraries
from bs4 import BeautifulSoup
import requests
import os
```

```python
#identify url we will be scraping
url = "https://www.rxlist.com/nexplanon-drug.htm#side_effects"
```

```python
#access data within url
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
html = soup.html
```

```python
print(soup.prettify())
```

```python
bc_type = soup.find('section').find('h1').text
bc_type
```

```python
bc_side_effects = soup.find('section', class_= "pageSection").find('div', class_= "pgContent").find_all('td')
bc_side_effects
```
