```python
#import libraries
from bs4 import BeautifulSoup
import requests
import os
```

```python
#identify url we will be scraping
url = "https://www.rxlist.com/ortho-tri-cyclen-drug.htm#indications"
```

```python
#access data within url
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
html = soup.html
```

```python
print(soup.prettify())
```

```python
bc_type = soup.find('section').find('h1').text
bc_type
```
