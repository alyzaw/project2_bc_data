```python
#import libraries
from bs4 import BeautifulSoup
import requests
import os
import pandas as pd
```

```python
#identify url we will be scraping
url = "https://www.rxlist.com/mirena-side-effects-drug-center.htm"
```

```python
#access data within url
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
html = soup.html
```

```python
print(soup.prettify())
```

```python
name = soup.find('section').find('h1').text
name
```

```python
list_symptoms = soup.find('div', attrs = {'class': 'pgContent'}).find('ul').text
symptoms = list_symptoms.replace("\n", " ")
symptoms
```

```python
d = [['Birth Control', name], ['Symptoms', symptoms]]
ETL_Data = pd.DataFrame(d)
ETL_Data
```
